```python
from tensorflow import keras
```

```python
from keras.api._v2.keras import activations
# dense1 = keras.layers.Dense(100,activation = 'sigmoid',input_shape=(28*28,),name='hidden')   # hidden
dense2 = keras.layers.Dense(10,activation = 'softmax',name='output')
model = keras.Sequential([
    keras.layers.Dense(100,activation = 'sigmoid',input_shape=(28*28,),name='hidden'),   # hidden
    dense2
    ] )
```

```python
model.summary()
```

```python
# compile
# fit  : epoch 5
```

```python
model.compile(loss='sparse_categorical_crossentropy',metrics='accuracy')
```

```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()
```

```python
x_train_scaled = x_train / 255.0
x_test_scaled = x_test / 255.0
```

```python
x_train_scaled_2d = x_train_scaled.reshape(-1,28*28)
x_test_scaled_2d = x_test_scaled.reshape(-1,28*28)
```

```python
history = model.fit(x_train_scaled_2d, y_train, epochs=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11), history.history['loss'], label='loss')
plt.plot(range(1,11), history.history['accuracy'], label='accuracy')
plt.legend()
plt.show()
```
