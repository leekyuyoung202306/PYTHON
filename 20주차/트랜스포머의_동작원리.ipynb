트랜스포커의 아키텍처
  - 인코더 디코더 구조를 기반
    - 인코더
      - 입력토큰의 시퀀스를 은닉상태 또는 문맥이라고 부르는 임베딩 벡터의 시퀀스로 변환
    - 디코더
      - 은닉상태를 사용해 출력토큰의 시퀀스를 한번에 하나씩 반복적으로 생성

image.png

트랜스포커 아키텍처는 크게 3가지 유형
  - 인코더 유형
    - 텍스트 시퀀스 입력을 풍부한 수치로표현, 텍스트분류나 개체명인식 같은 작업제 잘 맞고 BERT, RoBERTa, DistilBERT와 같은 BERT변종에 속함
    - 토큰에 대해 계산한 표현은 왼쪽과 오른쪽문맥에 따라서 달라지고 이를 양방향 어텐션이라고 부름
  - 디코더 유형
    - 시작텍스트가 주어지면 가장 가능성 있는 다음 단어를 반복해 예측하는 식으로 시퀀스를 자동 완성 GPT 계열의 모델
    - 토큰에 대해 계산한 표현은 오직 왼쪽 문맥에 따라서 달라진다, 코잘어텐션(causal attension), 자기회귀 어텐션(autoregressive attention)
  - 인코더-디코더 유형
    - 한 텍스트 시퀀스를 다른 시퀀스로 매핑하는 복잡한 모델링, 기계번역과 요약 작업에 적합, BART와 T5모델이 이 유형에 속함
