```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
import pandas as pd
```

```python
# !unzip '/content/drive/MyDrive/데이콘_신약개발/데이콘 신약데이터.zip'

train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
train_df = train.iloc[:,1:]
test_df = test.iloc[:,1:]
```

```python
test_df.head(2)
```

```python
save_path = './automl_MLM'
# label_columns = ['MLM','HLM']
predictor_MLM = TabularPredictor(
    label='MLM',
    path=save_path,
)
predictor_MLM.fit(
    train_data = train_df,
    time_limit=60*1
)

predictor_HLM = TabularPredictor(
    label='HLM',
    path=save_path,
)
predictor_HLM.fit(
    train_data = train_df,
    time_limit=60*1
)
```

```python
predict_MLM = predictor_MLM.predict(test_df)
predict_HLM = predictor_HLM.predict(test_df)
```

```python
# predict
df_submission = pd.read_csv("./sample_submission.csv")
df_submission["MLM"] = predict_MLM.values
df_submission["HLM"] = predict_HLM.values
df_submission.to_csv("automl_result.csv", index = False, encoding = "utf-8-sig")
```
