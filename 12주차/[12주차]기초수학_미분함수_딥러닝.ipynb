```python
from sympy import symbols  # x를 기호 변수로 바꿔주는 함수
def average(a,b):
  # 변화량을 구하기 위해서 두 수중 큰 수를 변수 m에 넣습니다.

  m = max(a,b)
  n = min(a,b)
  x = symbols('x')
  fx = x**2
  fb = fx.subs(x,m)
  fa = fx.subs(x,n)
  result = (fb-fa) / (b-a)
  return result
```

```python
average(1,4)
```

```python
# 미분의 기본 공식
# f(x) = x일때  f'(x) = 1
# f(x) = a에서 a가 상수일때 f'(x) = 0
# f(x) = ax에서 a가 상수일때 f'(x) = a
# f(x) = x**a에서 a가 자연수일때 f'(x) = ax**(a-1)
# f( g(x) )에서  f(x)와 g(x)가 미분가능  {f(g(x))}' = f'(g(x)) * g'(x)
```
