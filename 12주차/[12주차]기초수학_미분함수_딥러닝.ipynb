```python
from sympy import symbols  # x    
def average(a,b):
  #         m .

  m = max(a,b)
  n = min(a,b)
  x = symbols('x')
  fx = x**2
  fb = fx.subs(x,m)
  fa = fx.subs(x,n)
  result = (fb-fa) / (b-a)
  return result
```

```python
average(1,4)
```
