```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0,5.0,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python
# 가상의 데이터 생성
x = np.linspace(-5,5,100)
y = sigmoid(x) # 0 또는 1데 대한 확률계산
```

```python
# 임계값 설정
threshold = 0.5
```

```python
# 클래스 라벨링
predicted_labels = [1 if pred >= threshold else 0 for pred in y ]
predicted_labels
```

```python
plt.plot(x,y,label='sigmoid function')
plt.axhline(threshold,color='red',linestyle='--',label='threshold')
plt.scatter(x,predicted_labels,c='r',marker='x',label='predicted Labels')
plt.legend()
plt.xlabel('input feature')
plt.ylabel('predicted label')
plt.show()
```
