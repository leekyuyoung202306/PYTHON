3 x 3 크기의 넘파이배열을 생성, 모든 값을 2배로 증가시킨후 평균값을 계산(단 평균은 파이썬 리스트처럼 for문을 이용한다)

```python
temp = np.random.randint(0,100, size=(3,3))
temp
```

```python
temp *= 2
temp
```

```python
total = 0
for row in temp:
  for col in row:
    total += col
total
```

```python
total / 9
```

```python
ary = np.arange(15)
split_num = int(len(ary)*0.8)
train_data = ary[ : split_num]
test_data = ary[split_num : ]
split_num, train_data, test_data

test_data[:] = 100
test_data,  ary
```

리스트 슬라이스는 원본데이터를 잘라서 공유한다.
  - 독립적으로 존재(부분집합)하려면 복사를 해야 한다.

```python
ary = np.arange(15)
split_num = int(len(ary)*0.8)
train_data = ary[ : split_num]
test_data = ary[split_num : ].copy()
split_num, train_data, test_data

test_data[:] = 0
test_data,  ary
```
