5 x 5  크기의 랜덤값을 초기화된 2차원 리스트를 만들고 리스트의 모든 값에 100을 더하는 코드

```python
import random
```

```python
random.randint(0,100)
```

```python
python_list = [[random.randint(0,100)  for _ in range(5)]  for _ in range(5)          ]
python_list
```

```python
for row in range(5):
  for col in range(5):
    python_list[row][col] += 100
```

```python
type(python_list)
```

3 x 3 크기의 넘파이배열을 생성, 모든 값을 2배로 증가시킨후 평균값을 계산(단 평균은 파이썬 리스트처럼 for문을 이용한다)

```python
temp = np.random.randint(0,100, size=(3,3))
temp
```

```python
temp *= 2
temp
```

```python
total = 0
for row in temp:
  for col in row:
    total += col
total
```

```python
total / 9
```
