### 업종대분류의 데이터만 추출해서
  - 리스트로 변환
  - 키는 업종대분류가되고
  - 값은 횟수를 카운트 해 보자

```python
temp = [1,1,1,2,3,2,1,2,5,2,2,2]
count_dic = {}
for i in temp:
  if i not in count_dic:
    count_dic[i] = 1
  else:
    count_dic[i] += 1

count_dic
```

```python
df['업종대분류'].values
```

```python
temp = df['업종대분류'].values
count_dic = {}
for i in temp:
  if i not in count_dic:
    count_dic[i] = 1
  else:
    count_dic[i] += 1

count_dic
pd.DataFrame.from_dict(count_dic,orient='index', columns=['count'])
```

```python
df2 = pd.DataFrame(count_dic.values(),count_dic,columns=['count'])
df2.head()
```

```python
df2 = df2.sort_values(by=['count'],ascending=False)
df2_top3 = df2.head(3)
df2_top3
```
