```python
#  <div id='myId1'> 아기공룡 </div>
# <div id='myId1'> 아기공룡 </div>
# [<div class='myClass1'> 내 친구 </div>, <div class='myClass1'> 내 친구 </div>  ]
```

```python
with open('./sample02.html','rt',encoding='utf-8') as f:
  web_page = f.read()
  bs_object = bs4.BeautifulSoup(web_page,'html.parser')
  tag_div1 = bs_object.find('div', {'id':'myId1'})
  print(tag_div1.text)
  tag_div2 = bs_object.find('div', {'class':'myClass1'})
  print(tag_div2.text)
  tag_div3 = bs_object.find_all('div', {'class':'myClass1'})
  for tag in tag_div3:
    print(tag.text)
```

```python
with open('./sample02.html','rt',encoding='utf-8') as f:
  web_page = f.read()
  bs_object = bs4.BeautifulSoup(web_page,'html.parser')
  # a 태그 전부 추출하기
  a_list = bs_object.find_all('a')
  for atag in a_list:
    print(atag['href'])
```

- 링크주소는 http or https로 시작하는 것만 수집한다
- 수집한 데이터를 DataFrame형태로 보기좋게 가공한다

https://www.nate.com/ 주소에 있는 a 태그의 href 속성값만 추출

```python
nateurl = "https://www.nate.com"
nate_url = urllib.request.urlopen(nateurl)
bs_object = bs4.BeautifulSoup(nate_url,'html.parser')
atag_names = [];  atag_href = []
for atag in bs_object.find_all('a'):
  if atag['href'][:4] == 'http':
    atag_names.append(atag.text.strip())
    atag_href.append(atag['href'])
```

```python
import pandas as pd
df = pd.DataFrame({'이름' : atag_names, '링크주소':atag_href})
# df[:1]
# df.iloc[[0]]
df.loc[[0]]
```

저장
  - 데이터프레임객체.to_csv('저장할파일명.csv')

```python
df.to_csv("링크추출.csv",encoding='ms949')
```

```python
locale_code = ['11B10101','11D20501']
url = 'https://news.nate.com/weather?areaCode='
for local in locale_code:
  weather = urllib.request.urlopen(url+local)
  bs_object = bs4.BeautifulSoup(weather,'html.parser')
  weather_tags = bs_object.find('div', {'class':'right_today'})
  current = weather_tags.find('p', {'class':'celsius'}).text
  max_ = weather_tags.find('span',{'class':'maximum'}).text
  min_ = weather_tags.find('span',{'class':'minimum'}).text
  temp1 = weather_tags.find('em',{'class':'hide'}).text
  temp2 = weather_tags.find('em',{'class':'plus'}).text
  humidity = weather_tags.find('p',{'class':'humidity'}).text
  rainfall = weather_tags.find('p',{'class':'rainfall'}).text
  wind = weather_tags.find('p',{'class':'wind'}).text
  print(current , max_, min_, temp1, temp2, humidity,rainfall,wind)
```
