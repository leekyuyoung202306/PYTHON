```python
df = pd.read_csv('./singer2.csv',encoding='ms949')
df
```

```python
df[' '].str.replace(",","").astype(int)
```

```python
# for data in df[' '].values:
#   data.replace(',','')

new_list = [ data.replace(',','') for data in df[' '].values]

list(map(int, new_list))

plt.bar(df[''], list(map(int, new_list)))
plt.yscale('log')
```

```python
df[' _count'] = list(map(int, new_list))
df.head()
```

```python
df_sort = df.sort_values(by=' _count',ascending=False)
plt.bar(df_sort[''],df_sort[' _count'])
plt.yscale('log')
plt.show()
```

```python
plt.bar(df[''],df[''])
plt.show()
```

```python
fig, ax =  plt.subplots(1,2)
ax[0].bar(df_sort[''],df_sort[' _count'])
ax[0].set_yscale('log')
ax[0].set_title('youtube views')
ax[0].set_xticklabels(df_sort[''], rotation = 45)

df_sort_member = df.sort_values(by='',ascending=False)['']
df_sort_id = df.sort_values(by='',ascending=False)['']
ax[1].bar(df_sort_id,df_sort_member)
ax[1].set_title('members')
ax[1].set_xticklabels(df_sort_id, rotation = 45)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.bar(df_sort[''],df_sort[' _count'])
plt.yscale('log')
plt.title('views')
plt.xticks(rotation=45)


plt.subplot(1,2,2)
plt.bar(df_sort_id,df_sort_member)
plt.xticks(rotation=45)
plt.title('members')
plt.show()
```
