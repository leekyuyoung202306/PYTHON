Scatter

  - 회귀모델의 시각화(예측)

```python
x_data = np.random.rand(50)
y_data = np.random.rand(50)
size = (np.random.rand(50)*20)**2
colors = np.random.rand(50)
plt.scatter(x_data,y_data,s=size,alpha=0.5,c=colors,cmap='spring')
plt.colorbar()
plt.show()
```

CSV 파일 읽어서 시각화 해 보기
  - 유튜브 조회수 시각화

```python
df = pd.read_csv('./singer2.csv',encoding='ms949')
df
```

```python
df['유튜브 조회수'].str.replace(",","").astype(int)
```

```python
# for data in df['유튜브 조회수'].values:
#   data.replace(',','')

new_list = [ data.replace(',','') for data in df['유튜브 조회수'].values]

list(map(int, new_list))

plt.bar(df['아이디'], list(map(int, new_list)))
plt.yscale('log')
```

```python
df['유튜브 조회수_count'] = list(map(int, new_list))
df.head()
```

```python
df_sort = df.sort_values(by='유튜브 조회수_count',ascending=False)
plt.bar(df_sort['아이디'],df_sort['유튜브 조회수_count'])
plt.yscale('log')
plt.show()
```

```python
plt.bar(df['아이디'],df['인원'])
plt.show()
```

```python
fig, ax =  plt.subplots(1,2)
ax[0].bar(df_sort['아이디'],df_sort['유튜브 조회수_count'])
ax[0].set_yscale('log')
ax[0].set_title('youtube views')
ax[0].set_xticklabels(df_sort['아이디'], rotation = 45)

df_sort_member = df.sort_values(by='인원',ascending=False)['인원']
df_sort_id = df.sort_values(by='인원',ascending=False)['아이디']
ax[1].bar(df_sort_id,df_sort_member)
ax[1].set_title('members')
ax[1].set_xticklabels(df_sort_id, rotation = 45)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.bar(df_sort['아이디'],df_sort['유튜브 조회수_count'])
plt.yscale('log')
plt.title('views')
plt.xticks(rotation=45)


plt.subplot(1,2,2)
plt.bar(df_sort_id,df_sort_member)
plt.xticks(rotation=45)
plt.title('members')
plt.show()
```
