```python
from matplotlib import markers
import random
data1 = [random.randint(1,10) for _ in range(10)]
data2 = [random.randint(10,30) for _ in range(10)]
data3 = [random.randint(30,50) for _ in range(10)]
data4 = [random.randint(50,100) for _ in range(10)]
plt.plot(x,data1,label='a');   plt.plot(x,data2, linestyle='--',label='b');
plt.plot(x,data3,linestyle=':', label='c');   plt.plot(x,data4, linestyle='-.', label='d')
plt.legend()
plt.title('Test Graph')
plt.ylim(0,120)
plt.show()
```

```python
from matplotlib import markers
import random
data1 = [random.randint(1,10) for _ in range(10)]
data2 = [random.randint(10,30) for _ in range(10)]
data3 = [random.randint(30,50) for _ in range(10)]
data4 = [random.randint(50,100) for _ in range(10)]
plt.plot(x,data1,label='a');   plt.plot(x,data2, linestyle='--',label='b');
plt.plot(x,data3,linestyle=':', label='c');   plt.plot(x,data4, linestyle='-.', label='d')
plt.legend()
plt.title('Test Graph')
plt.ylim(0,120)
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```
